```python
import pandas as pd
```

```python
fear_greed_index=pd.read_csv('/content/fear_greed_index.csv')
historical_data=pd.read_csv('/content/historical_data.csv')
```

```python
fear_greed_index.head()
```

```python
historical_data.head()
```

```python
fear_greed_index['Date'] = pd.to_datetime(fear_greed_index['date']).dt.date
fear_greed_index['value'] = pd.to_numeric(fear_greed_index['value'], errors='coerce')
fear_greed_index['classification'] = fear_greed_index['classification'].str.strip().str.title()
```

```python
historical_data.columns = historical_data.columns.str.strip().str.lower().str.replace(' ', '_')
```

```python
historical_data['Date'] = pd.to_datetime(historical_data['timestamp_ist'], format='%d-%m-%Y %H:%M').dt.date
```

```python
numeric_columns = ['execution_price', 'size_tokens', 'size_usd', 'start_position', 'closed_pnl', 'fee']
for col in numeric_columns:
    historical_data[col] = pd.to_numeric(historical_data[col], errors='coerce')
```

```python
text_columns = ['account', 'coin', 'side', 'direction']
for col in text_columns:
    historical_data[col] = historical_data[col].astype(str).str.strip().str.upper()
```

```python
historical_data.drop_duplicates(inplace=True)
fear_greed_index.drop_duplicates(inplace=True)
```

```python
historical_data.info()
```

```python
merged = pd.merge(historical_data, fear_greed_index[['Date', 'value', 'classification']], on='Date', how='left')
```

```python
merged.head()
```

```python
merged.to_csv('merged.csv', index=False)
```
